```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("age_income.csv")
df.head()
```

```python
plt.scatter(df.Age, df["Income($)"])
```

```python
km = KMeans(n_clusters = 3)
cluster = km.fit_predict(df[["Age", "Income($)"]])
print(cluster)
df['cluster'] = cluster
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1.Age, df1["Income($)"])
plt.scatter(df2.Age, df2["Income($)"])
plt.scatter(df3.Age, df3["Income($)"])
```

```python
sdf1 = df[df.cluster == 0]
sdf2 = df[df.cluster == 1]
sdf3 = df[df.cluster == 2]

# plt.scatter(sdf1.Age, sdf1["Income($)"])
# plt.scatter(sdf2.Age, sdf2["Income($)"])
# plt.scatter(sdf3.Age, sdf3["Income($)"])

plt.scatter(sdf1["Income($)"], sdf1.Age)
plt.scatter(sdf2["Income($)"], sdf2.Age)
plt.scatter(sdf3["Income($)"], sdf3.Age)
```

```python
scaler = MinMaxScaler()
df.Age = scaler.fit_transform(df[["Age"]])
df["Income($)"] = scaler.fit_transform(df[["Income($)"]])
df.head()
```
