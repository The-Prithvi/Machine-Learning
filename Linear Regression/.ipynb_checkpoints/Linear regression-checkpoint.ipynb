```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv("./gold_rate.csv")
x = df.Year
y = df.Rate
plt.scatter(x, y, marker = "+", color = 'r')
reg = linear_model.LinearRegression()
# reg.fit(df[["Year"]], df.Rate)     ## fit(xaxis, yaxis)
reg.fit(df[["Year"]], df.Rate)     ## fit(xaxis, yaxis)
# plt.plot(df.Year, reg.predict(df[['Year']]))
plt.plot(df.Year, reg.predict(df[["Year"]]))
print(reg.predict(np.array(2000).reshape(-1, 1)))
plt.show()
```

```python
df2 = pd.read_csv("./predictions.csv")
rate = reg.predict(df2)
df2["Rate"] = rate
df2.to_csv("./predictions.csv")
```
