```python
import pandas as pd
```

```python
data = pd.read_csv("titanic.csv")
data
df = data.drop(["Name", "SibSp", "Parch", "Ticket", "Cabin", "Embarked", "PassengerId"], axis = "columns")
df.Age.fillna(df.Age.mean(), inplace = True)
df
```

```python
dummy = pd.get_dummies(df.Sex)
dummy = dummy.astype(int)  # converts true and false into 1 and 0
dummy.head()
```

```python
df = pd.concat([df, dummy], axis = "columns")
df
```

```python
df = df.drop(["Sex"], axis = "columns")
df
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

x, y = df.drop(["Survived"], axis = "columns"), df.Survived

# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.02, random_state = 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.02)

model = GaussianNB()
model.fit(x_train, y_train)
print(model.score(x_test, y_test))
y_test, model.predict(x_test)
```
